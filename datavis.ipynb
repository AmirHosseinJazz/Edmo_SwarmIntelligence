```python
import csv



def count_occurrences(csv_file):
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Read the header row

        # Create a dictionary to store the counts for each column
        counts = {column: {} for column in header}

        # Iterate over each row in the CSV file
        for row in reader:
            for index, item in enumerate(row):
                column = header[index]
                column_counts = counts[column]

                # Increment the count for the current item in the column
                column_counts[item] = column_counts.get(item, 0) + 1

    return counts

# Usage example
csv_file = "samples.csv"
result = count_occurrences(csv_file)

# Display the counts
for column, counts in result.items():
    print(f"Column '{column}':")
    for item, count in counts.items():
        print(f"- {item}: {count} occurrences")
    print()
```

```python
import csv
import matplotlib.pyplot as plt

def count_occurrences(csv_file):
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        header = next(reader)  # Read the header row

        # Create a dictionary to store the counts for each column
        counts = {column: {} for column in header}

        # Iterate over each row in the CSV file
        for row in reader:
            for index, item in enumerate(row):
                column = header[index]
                column_counts = counts[column]

                # Increment the count for the current item in the column
                column_counts[item] = column_counts.get(item, 0) + 1

    return counts

def create_bar_charts(counts):
    for column, data in counts.items():
        items = data.keys()
        counts = data.values()

        # Create a bar chart
        plt.bar(items, counts)
        plt.xlabel(column)
        plt.ylabel('Occurrences')
        plt.title(f'Occurrences of items in {column}')
        plt.show()

# Usage example
result = count_occurrences(csv_file)

# Create bar charts
create_bar_charts(result)
```
