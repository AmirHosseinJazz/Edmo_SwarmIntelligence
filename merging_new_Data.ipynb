```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('NewParameterSamples.csv')
```

```python
speed=pd.read_csv('speed_data.csv')
```

```python
speed.columns=['Timestamp','Speed']
```

```python
t=pd.merge(speed,data,on=['Timestamp'],how='outer')
```

```python
t=t.sort_values(by='Timestamp')
```

```python
for i in t.columns:
    if i!='Timestamp' and i!='Speed':
        t[i]=t[i].fillna(method='ffill')
```

```python
t=t.dropna(subset=['Snelheid'])
```

```python
t['Speed']=t['Speed'].fillna(method='bfill')
```

```python
results=t.groupby(['Snelheid', 'Omvang1', 'Positie1', 'Omvang2','Positie2', 'Relatie']).agg({'Speed':np.mean})
```

```python
Final=results.reset_index()
```
