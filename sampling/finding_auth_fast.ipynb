```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('NewParameterSamples_0-200.csv')
speed=pd.read_csv('speed_data_new_0-200.csv')
speed=speed[speed['Time']>='11:46:00.000']
speed.columns=['Timestamp','Speed']
t=pd.merge(speed,data,on=['Timestamp'],how='outer')
t=t.sort_values(by='Timestamp')
for i in t.columns:
    if i!='Timestamp' and i!='Speed':
        t[i]=t[i].fillna(method='ffill')
t=t.dropna(subset=['Snelheid'])        
t['Speed']=t['Speed'].fillna(method='bfill')
```

```python
for index,row in t.iterrows():
    f=str(row['Snelheid'])+','+str(row['Omvang1'])+','+str(row['Positie1'])+','+str(row['Omvang2'])+','+str(row['Positie2'])+','+str(row['Relatie'])
    t.loc[index,'params']=f
```

```python
for j in t.params.unique():
    speeds=(t[t['params']==j].Speed)
    atleastOne=False
    atleastTwo=False
    for k in speeds:
        if k>2:
            if atleastOne:
                atleastTwo=True
            else:
                atleastOne=True
    if atleastTwo:
        print(j)
        print(speeds)
```
